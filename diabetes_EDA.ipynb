```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('diabetes_medicines__data.csv')
data
```

```python
brands = []
for p in data['product'] :
    if p.startswith('Dr') :
        temp_str = p.split(' ')[0]+p.split(' ')[1]
        brands.append(temp_str)
    else :
        temp_str = p.split(' ')[0]
        brands.append(temp_str)

brands
```

```python
data['brand'] = brands
```

```python
data
```

```python
# data.to_csv('diabetes_medicines__data_final.csv', index=False)
```

```python
data['brand'].unique()
```

```python
brand_wise_rating = data.groupby('brand')['rating'].mean().sort_values()[-10:]
plt.barh(brand_wise_rating.keys(), brand_wise_rating)
plt.xlabel('rating')
plt.show()
```

```python
brand_wise_count = data.groupby('brand')['rating'].count().sort_values()[-10:]
plt.barh(brand_wise_count.keys(), brand_wise_count)
plt.xlabel('number of products')
plt.show()
```

### if we consider that more ratings is directly proportional to sale...

```python
brand_wise_rating_count = data.groupby('brand')['number of ratings'].sum().sort_values()[-10:]
plt.barh(brand_wise_rating_count.keys(), brand_wise_rating_count)
plt.xlabel('rating count')
plt.show()
```

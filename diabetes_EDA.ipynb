```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('diabetes_medicines__data.csv')
data
```

```python
brands = []
for p in data['product'] :
    if p.startswith('Dr') :
        temp_str = p.split(' ')[0]+p.split(' ')[1]
        brands.append(temp_str)
    else :
        temp_str = p.split(' ')[0]
        brands.append(temp_str)

brands
```

```python
data['brand'] = brands
```
